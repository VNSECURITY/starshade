```python
import pandas
import numpy
import scipy
import tensorflow as tf
import numpy as np
```

```python
"""
CSIC data: http://www.isi.csic.es/dataset/
"""
csic_data = pandas.read_csv("output_http_csic_2010_weka_with_duplications_utf8_escd_v02_full.csv",error_bad_lines=False)
csic_test = pandas.read_csv("output_http_csic_2010_weka_with_duplications_utf8_escd_v02_norm_test.csv",error_bad_lines=False)
```

```python
"""
TODO: use a better vector ^_^
"""
CS = "0123456789abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ@#?"
CSrev = {}
for i,c in enumerate(CS):
    CSrev[ord(c)] = i+1
def make_feature_vector(sample):
    features = [0] * (len(CS)+1)
    for c in str(sample['payload']):
        features[CSrev.get(ord(c),0)] += 1
    return map(lambda x: np.tanh(x), features), [0 if sample['label']=='anom' else 1]
```

```python
train_data = [make_feature_vector(x) for index,x in csic_data.iterrows()]
train_x = [x[0] for x in train_data]
train_y = [x[1] for x in train_data]

test_data = [make_feature_vector(x) for index,x in csic_test.iterrows()]
test_x = [x[0] for x in test_data]
test_y = [x[1] for x in test_data]
```

```python
inp_x = tf.placeholder("float", [None, len(CS)+1])
inp_y = tf.placeholder("float", [None, 1])

def nn(x, weights, biases):
    for i in range(1):
        x = tf.nn.tanh(tf.add(tf.matmul(x, weights['w%d'%(i)]), biases['b%d'%(i)]))
    return tf.matmul(x, weights['out']) + biases['out']

weights = {
    'w0': tf.Variable(tf.random_normal([len(CS)+1, 10])),
    'out': tf.Variable(tf.random_normal([10, 1]))
}
biases = {
    'b0': tf.Variable(tf.random_normal([10])),
    'out': tf.Variable(tf.random_normal([1]))
}
```

```python
net = nn(inp_x, weights, biases)
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(net, inp_y))
optimizer = tf.train.AdamOptimizer(learning_rate=0.001).minimize(cost)
```

```python
init = tf.initialize_all_variables()
batch_size = 100
with tf.Session() as sess:
    sess.run(init)
    for epoch in range(10):
        avg_cost = 0.
        total_batch = int(len(train_x)/batch_size)
        for i in range(int(len(train_x)/batch_size)):
            fd = {inp_x: train_x[batch_size*i:batch_size*(i+1)], inp_y: train_y[batch_size*i:batch_size*(i+1)]}
            _, c = sess.run([optimizer, cost], feed_dict=fd)
            avg_cost += c / total_batch
        if epoch % 1 == 0:
            print("Epoch:", '%04d' % (epoch+1), "cost=", \
                "{:.9f}".format(avg_cost))
    tf.train.Saver().save(sess, 'model.bin')
```

```python
with tf.Session() as sess:
    tf.train.Saver().restore(sess, 'model.bin')
    correct_prediction = tf.equal(tf.argmax(net, 1), tf.argmax(inp_y, 1))
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
    print accuracy.eval({inp_x: train_x, inp_y: train_y})
```
